```python
import requests
from bs4 import BeautifulSoup
import re
```

```python
url = 'http://kcna.kp/kcna.user.home.retrieveHomeInfoList.kcmsf'
raw_html = requests.get(url).content
soup_doc = BeautifulSoup(raw_html, "html.parser")
print(soup_doc.prettify)
```

```python
soup_doc.find_all()
```

```python
articles = soup_doc.find_all('h3')

data_list = []
for article in articles:
    title = article.find_all(class_='titlebet')[0].text
    on_click= article.find_all('a')[0].get('onclick')
    data_dictionary={'title': title, 'on_click':on_click}
    data_list.append(data_dictionary)
```

```python
import pandas as pd 
```

```python
df = pd.DataFrame(data_list)
```

```python
df
```

```python
df['id'] = df['on_click'].str.extract(r"\(\W([A-Z][A-Z]\d+)")
```

```python
df.to_csv('nk-news.csv')
```
