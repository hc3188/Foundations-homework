````python
my_url = "http://www.mineral.k12.nv.us/pages/School_Board_Minutes"
raw_html = requests.get(my_url).content
soup_doc = BeautifulSoup(raw_html, "html.parser")
print(soup_doc.prettify)
````

````python
a_tags = driver.find_elements_by_tag_name('a')[17:76]

#print(a_tags)

for tag in a_tags:
    print(tag.text)
````

````python
list_of_links = []
for tag in a_tags:
    date = tag.text
    links = tag.get_attribute('href')
    dictionary={'date': date, 'links': links}
    list_of_links.append(dictionary)
                
````

````python
print(list_of_links)
````

````python
import pandas as pd 
````

````python
df = pd.DataFrame(list_of_links)
````

````python
df
````

````python
options = Options()
options.add_experimental_option('prefs',  {
    "plugins.always_open_pdf_externally": True
})
for link in list_of_links:
    driver = webdriver.Chrome(options=options)
    driver.get(link)
````
