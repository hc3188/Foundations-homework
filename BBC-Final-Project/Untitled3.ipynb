```python
import pandas as pd 
import numpy as np
```

```python
df_movie_by_country = pd.read_csv('real-last-dataframe.csv')
df_movie_by_country

```

```python
df1 = df_movie_by_country.groupby('Country')['Total Worldwide Box Office'].value_counts().reset_index(name='No. of Movies')
df1["string"] = df1["Total Worldwide Box Office"] + ": " + df1["No. of Movies"].map(str)+ np.where(df1["No. of Movies"]>1, ' movies', ' movie')
df1
```

```python
article = df1.groupby('Country')['string'].apply(lambda x: "<div id='Total Worldwide Box Office'><P>%s</P></div>" % '</p><p> '.join(x)).reset_index(name='properties.article')
article
```

```python
headline = df_movie_by_country["Country"].value_counts().reset_index(name='properties.headline')
headline = headline.rename(columns={'index': 'Country'})
```

```python
output = article.merge(headline, how='left', on='Country')
output
```

```python
output['properties.color'] = "#35476E"
output
```

```python
output
```

```python
output.to_csv('output')
```
