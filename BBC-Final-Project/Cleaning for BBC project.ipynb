```python
import pandas as pd 
```

```python
movie_critic_df = pd.read_csv('movie_critic_list')
```

```python
movie_critic_df.dtypes
```

```python
movie_critic_df['country_string'] = movie_critic_df['critic_country'].astype(str)
```

```python
movie_critic_df
```

```python
regex_for_critic_country= r'\W(.*)\W'
movie_critic_df['country_string'] =movie_critic_df['country_string'].str.extract(regex_for_critic_country)



```

```python
movie_critic_df['critic_country'] = movie_critic_df['country_string']
```

```python
movie_critic_df.drop('country_string',axis=1, inplace = True)
```

```python
movie_critic_df.to_csv('cleaned_movie_critic_list')
```
