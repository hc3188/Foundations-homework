```python
final_list = []
for item in updated_titles_list:
    url= 'https://www.imdb.com/'+item
    raw_html = requests.get(url).content
    soup_doc = BeautifulSoup(raw_html, "html.parser")
    title = soup_doc.find('div', class_='title_wrapper').find('h1').text.split('(')
    details_box=soup_doc.find('div',class_="article", id='titleDetails')
    txt_boxes= details_box.find_all('div', class_="txt-block")
    for box in txt_boxes:
        inner_list=[]
        h4= box.find('h4')
        try:
            if h4.text == 'Country:':
                a_tags = h4.find_next_siblings('a')
                inner_list.append(title)
                for tag in a_tags:
                    tags = tag.text
                    #print(tags)
                    inner_list.append(tags)
                
                #print(inner_list)
                final_list.append(inner_list)
    #final_list.append
            else:
                pass
        except:
            pass
```
